# **TEXT CLASSIFICATION**
USING SENTIMENT ANALYSIS WITH PIPELINE FUNCITON IN TRANSFORMERS LIBRARY

```python
from transformers import pipeline
demo = pipeline("sentiment-analysis")
demo(['i love you',
     'i hate you',
     'i love you but i am not in love with you'])
```

**USING ZERO SHOT CLASSIFICATION USING PIPELINE FUNCTION IN TRANSFORMERS LIBRARY**

We’ll start by tackling a more challenging task where we need to classify texts that haven’t been labelled.

This is a common scenario in real-world projects because annotating text is usually time-consuming and requires domain expertise.

For this use case, the zero-shot-classification pipeline is very powerful: it allows you to specify which labels to use for the classification, so you don’t have to rely on the labels of the pretrained model.

You’ve already seen how the model can classify a sentence as positive or negative using those two labels — but it can also classify the text using any other set of labels you like

```python
div = pipeline("zero-shot-classification")
div('tell me about something that i donot know',
    candidate_labels=['Behavior','Happy','Sad'])
```

# **Text Generation**
**USING TEXT GENERATION FUNCTION WITH PIPELINE FUNCTION IN TRASFORMERS LIBRARY**

The main idea here is that you provide a prompt and the model will auto-complete it by generating the remaining text. This is similar to the predictive text feature that is found on many phones. Text generation involves randomness, so it’s normal if you don’t get the same results as shown below.

```python
sample = pipeline('text-generation')
sample('tell me about')
```

```python
generator = pipeline('text-generation', model='distilgpt2')
generator('it has been great meeting you',
max_length=30,
num_return_sequences=3)
```

# **Bias and Limitations**

```python
from transformers import pipeline
unmasked=pipeline('fill-mask', model='bert-base-uncased')
out = unmasked('This man works as [MASK].')
print([i['token_str'] for i in out])
out = unmasked('This woman works as [MASK].')
print([i['token_str'] for i in out])
```
